```python
## Build histograms
InputFolder = '/Users/christine/Documents/Data/DynamoLINC/FLIM/FLIM_240920/'

ls = os.listdir(InputFolder+'Masks')


minval = 1.3
maxval = 3.2
b=25

cond = []
hist = []


for f in ls:
    condition = f.split('-')[0]
    cond.append(condition)
    mask = io.imread(InputFolder+'Masks/'+f).astype(float)/255
    img_FLIM = io.imread(InputFolder+'FLIMdata/'+f).astype(float)
    
    FLIM = img_FLIM*mask
    
    fig,axs = plt.subplots(1,4,figsize=(16,4))
    fig.suptitle(f)
    axs[0].axis('off')
    axs[1].axis('off')
    axs[2].axis('off')
    msk=axs[0].imshow(mask)
    axs[0].set_title('Mask')
    or_flim = axs[1].imshow(img_FLIM)
    axs[1].set_title('FLIM')
    flim = axs[2].imshow(FLIM)
    axs[2].set_title('mask & FLIM')
    fig.colorbar(flim,ax=axs[2],shrink=0.7)
    fig.colorbar(or_flim,ax=axs[1],shrink=0.7)
    fig.colorbar(msk,ax=axs[0],shrink=0.7)
    
    
    histogram, bin_edges = np.histogram(FLIM, bins=b, range=(minval, maxval))
    hist.append(histogram)
    axs[3].plot(bin_edges[0:-1], histogram)
    axs[3].set_xlabel('Life time (ns)')
    plt.savefig(InputFolder+'Results/'+f)

DF = pd.DataFrame()
DF['Condition'] = cond[:]
DF['Histograms'] = hist[:]
```

```python
DF_10 = DF[DF['Condition']=='10um']
DF_100 = DF[DF['Condition']=='100um']
DF_square = DF[DF['Condition']=='square']

h10 = np.zeros(b)
h100 = np.zeros(b)
hsquare = DF_square['Histograms'].iloc[0]

for i in range(len(DF_10)):
    h10 = h10 + DF_10['Histograms'].iloc[i]

for i in range(len(DF_100)):
    h100 = h100 + DF_100['Histograms'].iloc[i]
    
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(bin_edges[0:-1],h10, label='10um')
ax.plot(bin_edges[0:-1],h100, label='100um')
ax.plot(bin_edges[0:-1],hsquare, label='square')
ax.set_xlabel('Life time (ns)')
plt.legend()
plt.savefig(InputFolder+'/Results/Histograms.jpg')
        
        
        
        
    
```
